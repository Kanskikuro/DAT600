```python
import heapsort
import insertion_sort
import mergesort 
import quicksort 
```

```python
import time
import random
import matplotlib.pyplot as plt

input_sizes = [10, 50, 100, 500, 1000]
step_counters = {"Insertion Sort": [], "Merge Sort": [], "Heap Sort": [], "Quicksort": []}

for n in input_sizes:
    # Generate random array of size n
    test_array = [random.randint(-1000, 1000) for _ in range(n)]
    
    # Insertion Sort
    arr = test_array[:]
    start_time = time.time()
    _, insertion_step_count = insertion_sort.insert_sort(arr)
    step_counters["Insertion Sort"].append(insertion_step_count)
    
    # Merge Sort
    arr = test_array[:]
    start_time = time.time()
    _, merge_step_count = mergesort.mergesort(arr)
    step_counters["Merge Sort"].append(merge_step_count)
    
    # Heap Sort
    arr = test_array[:]
    start_time = time.time()
    _, heap_step_count = heapsort.heapsort(arr)
    step_counters["Heap Sort"].append(heap_step_count)
    
    # Quicksort
    arr = test_array[:]
    start_time = time.time()
    _, quick_step_count = quicksort.quicksort(arr)
    step_counters["Quicksort"].append(quick_step_count)

#Print the step counts
print("Insertion Sort Step Counts: ", insertion_step_count)
print("Merge Sort Step Counts: ", merge_step_count)
print("Heap Sort Step Counts: ", heap_step_count)
print("Quicksort Step Counts: ", quick_step_count)

# Plot the results
plt.figure(figsize=(12, 8))
for algorithm, step_count in step_counters.items():
    plt.plot(input_sizes, step_count, label=algorithm)

plt.title("Execution Time of Sorting Algorithms")
plt.xlabel("Input Size (n)")
plt.ylabel("Step Count")
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data from the C program (without optimization)
c_lengths = [100, 1000, 5000, 10000, 20000]
c_times_no_opt = [0.000007, 0.000414, 0.010689, 0.040088, 0.158119]

# Updated data from the C program (with -O3 optimization)
c_times_o3 = [0.000007, 0.000191, 0.001823, 0.008628, 0.030428]

# Data from the Go program
go_lengths = [100, 1000, 5000, 10000, 20000]
go_times = [0.000007, 0.000361, 0.005154, 0.020659, 0.081810]

# Plotting the data
plt.figure(figsize=(8, 6))
plt.plot(c_lengths, c_times_no_opt, label='C (No Optimization)', marker='o', linestyle='-', color='b')
plt.plot(c_lengths, c_times_o3, label='C (-O3 Optimization)', marker='^', linestyle='-', color='g')
plt.plot(go_lengths, go_times, label='Go', marker='s', linestyle='-', color='r')

# Labeling the graph
plt.xlabel('Array Length')
plt.ylabel('Time (seconds)')
plt.title('Insertion Sort: C (No Optimization) vs C (-O3) vs Go')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
