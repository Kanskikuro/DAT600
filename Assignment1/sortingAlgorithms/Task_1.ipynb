```python
import heapsort
import insertion_sort
import mergesort 
import quicksort 
```

```python
import time
import random
import matplotlib.pyplot as plt

input_sizes = [10, 50, 100, 500, 1000, 10000, 20000]
execution_times = {"Insertion Sort": [], "Merge Sort": [], "Heap Sort": [], "Quicksort": []}

for n in input_sizes:
    # Generate random array of size n
    test_array = [random.randint(1, 1000) for _ in range(n)]
    
    # Insertion Sort
    arr = test_array[:]
    start_time = time.time()
    insertion_sort.insert_sort(arr)
    execution_times["Insertion Sort"].append(time.time() - start_time)
    
    # Merge Sort
    arr = test_array[:]
    start_time = time.time()
    mergesort.mergesort(arr)
    execution_times["Merge Sort"].append(time.time() - start_time)
    
    # Heap Sort
    arr = test_array[:]
    start_time = time.time()
    heapsort.heapsort(arr)
    execution_times["Heap Sort"].append(time.time() - start_time)
    
    # Quicksort
    arr = test_array[:]
    start_time = time.time()
    quicksort.quicksort(arr)
    execution_times["Quicksort"].append(time.time() - start_time)

# Plot the results
plt.figure(figsize=(12, 8))
for algorithm, times in execution_times.items():
    plt.plot(input_sizes, times, label=algorithm)

plt.title("Execution Time of Sorting Algorithms")
plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (seconds)")
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data from the C program (without optimization)
c_lengths = [100, 1000, 5000, 10000, 20000]
c_times_no_opt = [0.000007, 0.000414, 0.010689, 0.040088, 0.158119]

# Updated data from the C program (with -O3 optimization)
c_times_o3 = [0.000007, 0.000191, 0.001823, 0.008628, 0.030428]

# Data from the Go program
go_lengths = [100, 1000, 5000, 10000, 20000]
go_times = [0.000007, 0.000361, 0.005154, 0.020659, 0.081810]

# Plotting the data
plt.figure(figsize=(8, 6))
plt.plot(c_lengths, c_times_no_opt, label='C (No Optimization)', marker='o', linestyle='-', color='b')
plt.plot(c_lengths, c_times_o3, label='C (-O3 Optimization)', marker='^', linestyle='-', color='g')
plt.plot(go_lengths, go_times, label='Go', marker='s', linestyle='-', color='r')

# Labeling the graph
plt.xlabel('Array Length')
plt.ylabel('Time (seconds)')
plt.title('Insertion Sort: C (No Optimization) vs C (-O3) vs Go')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
